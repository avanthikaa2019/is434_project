```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

creators = ['glossyco','gucci','burberry','LouisVuitton','balenciaga','dolcegabbana','RebeccaMinkoff','KarlLagerfeld','bnv_me','nike','RTFKTstudios','Overpriced_NFT','auroboros_ltd','republiqec','the_fab_ric_ant']
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for user in creators:
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('(NFT OR metaverse) from:'+ user + ' lang:en -is:retweet since:2021-01-01 until:2022-03-01').get_items()): 
        if i>1000: #number of tweets you want to scrape
            break

        #declare the attributes to be returned                                        
        tweets_list1.append([tweet.user.username, tweet.date, tweet.id, tweet.content, 
                             tweet.likeCount, tweet.retweetCount, tweet.replyCount, tweet.quoteCount, tweet.mentionedUsers]) 

    
# Creating a dataframe from the tweets list above 
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Username','Datetime', 'Tweet Id', 'Text', 'Likes', 
                                                 'Retweets', 'Replies',  'Quotes', 'Mentioned'])
                                                        
tweets_df1
```
