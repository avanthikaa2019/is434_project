# SNSCRAPE

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
creators = ['gucci','adidas','adidasUS','burberry','LouisVuitton','balenciaga','dolcegabbana','RebeccaMinkoff',
            'KarlLagerfeld','bnv_me','nike','RTFKTstudios','Overpriced_NFT','auroboros_ltd','republiqec',
            'the_fab_ric_ant','SpaceRunnersNFT','MetaTriads','MAVIONworld','SohoKidsNFT','onclickcloset',
            'SCLadiesNFTs','mag_nezyum','web3fashion','DIGITALAX_','prada','balmain','ray_ban','superplastic',
           'givenchy','Bulgariofficial','MaisonValentino','fendi','jimmychoo','Guerlain','MetaBirkins','RalphLauren',
           'dressxcom','tribute_brand','dematerialised','XRCouture','carlingscom']

print("number of creators scraped:",len(creators))

print("query: NFT OR metaverse, english language, not retweet, from 1 Jan 2021 to 1 Mar 2022")

creator_tweets = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for user in creators:
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper('(NFT OR metaverse) from:'+ user + ' lang:en -is:retweet since:2021-01-01 until:2022-03-01').get_items()): 
        if i>2000: #number of tweets you want to scrape
            break

        #declare the attributes to be returned                                        
        creator_tweets.append([tweet.user.username, tweet.date, tweet.id, tweet.content, 
                             tweet.likeCount, tweet.retweetCount, tweet.replyCount, tweet.quoteCount, tweet.mentionedUsers]) 

    
creator_tweets_df = pd.DataFrame(creator_tweets, columns=['Username','Datetime', 'Tweet Id', 'Text', 'Likes', 
                                                 'Retweets', 'Replies',  'Quotes', 'Mentioned'])

print(creator_tweets_df.shape)
print(creator_tweets_df['Username'].value_counts())
```
