## word clean up set up

```python
import numpy as np
import re
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import matplotlib.pyplot as plt

stop_list = stopwords.words('english')
other_words = ['amp']
stop_list.extend(other_words)

def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = re.sub("'", "", temp)
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp) #keep hashtags?
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = temp.split() 
    temp = [w for w in temp if not w in stop_list]
    temp = " ".join(word for word in temp)
    return temp

print("done")
```

## tweet text processed

```python
media_text_list = media_tweets_csv['Text'].tolist()
media_results = [clean_tweet(tw) for tw in media_text_list]
print('done')
```

```python
creator_text_list = creator_tweets_csv['Text'].tolist()
creator_results = [clean_tweet(tw) for tw in creator_text_list]
print('done')
```

```python
tokenizer = RegexpTokenizer(r'\w+')
porter_stemmer = PorterStemmer()

media_words = []
for tweet_text in media_results:
    tokens = tokenizer.tokenize(tweet_text)
    media_words.extend(tokens)

print(len(media_words))

media_stemmed = []
for w in media_words:
    media_stemmed.append(porter_stemmer.stem(w))
    
print(len(media_stemmed))
```

```python
tokenizer = RegexpTokenizer(r'\w+')
porter_stemmer = PorterStemmer()

creator_words = []
for tweet_text in creator_results:
    tokens = tokenizer.tokenize(tweet_text)
    creator_words.extend(tokens)

print(len(creator_words))

creator_stemmed = []
for w in creator_words:
    creator_stemmed.append(porter_stemmer.stem(w))
    
print(len(creator_stemmed))
```

```python
import collections
media_frequency = collections.Counter(media_stemmed)
media_dic = dict(media_frequency)
print(len(media_dic))

a1_sorted_keys = sorted(media_dic, key=media_dic.get, reverse=True)
for r in a1_sorted_keys:
    print(r, media_dic[r])
```

```python
creator_frequency = collections.Counter(creator_words)
creator_dic = dict(creator_frequency)
print(len(creator_dic))

a2_sorted_keys = sorted(creator_dic, key=creator_dic.get, reverse=True)
for r in a2_sorted_keys:
    print(r, creator_dic[r])
```
